```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
#生成n维对称正定矩阵A及n维向量b
def SPDM(n):
    M = np.random.randn(n,n)#生成n维随机矩阵
    C = np.dot(M,M.T)#生成半正定矩阵
    D = C + C.T#生成对称半正定矩阵
    A = D + np.eye(n) #生成n维对称正定矩阵
    A = np.round(A,3)#保留三位有效数字
    b = np.random.randn(n)#生成n维列向量
    b = np.round(b,3)
    return A,b

#共轭梯度法
def CGD():
    n = int(input("请输入矩阵A及向量b的维度n："))
    A,b = SPDM(n)
    print("对称正定矩阵A为：\n",A)
    print("向量b为：\n",b)
    epsilon = float(input("请输入精度epsilon(epsilon>0):"))
    print("精度epsilon为：\n",epsilon)
    x = np.random.randn(n)#随机生成初值x0
    d = b - np.dot(A,x.T)#d0
    r_k = d
    r_k1 = d
    i = 0
    e_norm = []#用于储存残差向量组的范数
    x_ite = []#用于储存迭代过程中的x
    x_ite.append(x)
    e_norm.append(np.linalg.norm(r_k1))
    while(np.linalg.norm(r_k1) / np.linalg.norm(b) > epsilon):
        r_k = r_k1
        alpha = (np.linalg.norm(r_k) ** 2) / (np.dot(d,np.dot(A,d)))
        x = x + alpha * d
        x_ite.append(x)
        r_k1 = b - np.dot(A,x)
        e_norm.append(np.linalg.norm(r_k1))
        beta = (np.linalg.norm(r_k1) ** 2) / (np.linalg.norm(r_k) ** 2)
        d = r_k1 + beta * d
        i += 1
    print("迭代总次数为：",i)
    print("求得的解x为：\n",x)
    print("残差向量为：\n",r_k1)
    print("x的更替过程为：\n",x_ite)
    x_l = np.arange(i+1)
    fig = plt.figure(figsize = (20,10),dpi = 80)
    plt.plot(x_l,e_norm)
    plt.xlabel("迭代次数",fontproperties = "STSong")
    plt.ylabel("残差向量的二范数",fontproperties = "STSong")
    plt.title("残差向量的二范数随迭代次数的变化",fontproperties = "STSong")
    plt.show()
```

```python
CGD()
```
